```python
import pandas as pd

data_2010 = pd.read_csv('ACSST1Y2010.S1701-Data.csv')
data_2011 = pd.read_csv('ACSST1Y2011.S1701-Data.csv')
data_2012 = pd.read_csv('ACSST1Y2012.S1701-Data.csv')
data_2013 = pd.read_csv('ACSST1Y2013.S1701-Data.csv')
data_2014 = pd.read_csv('ACSST1Y2014.S1701-Data.csv')
data_2015 = pd.read_csv('ACSST1Y2015.S1701-Data.csv')
data_2016 = pd.read_csv('ACSST1Y2016.S1701-Data.csv')
data_2017 = pd.read_csv('ACSST1Y2017.S1701-Data.csv')
data_2018 = pd.read_csv('ACSST1Y2018.S1701-Data.csv')
data_2019 = pd.read_csv('ACSST1Y2019.S1701-Data.csv')
data_2021 = pd.read_csv('ACSST1Y2021.S1701-Data.csv')
data_2022 = pd.read_csv('ACSST1Y2022.S1701-Data.csv')
data_2023 = pd.read_csv('ACSST1Y2023.S1701-Data.csv')


datasets = [
    (data_2010, 2010), (data_2011, 2011), (data_2012, 2012),
    (data_2013, 2013), (data_2014, 2014), (data_2015, 2015),
    (data_2016, 2016), (data_2017, 2017), (data_2018, 2018),
    (data_2019, 2019), (data_2021, 2021), (data_2022, 2022), (data_2023, 2023)
]

#Relevant Columns to our Research Question
columns = [
    'GEO_ID', 'NAME', 'S1701_C01_001E',#Total population
    'S1701_C02_001E', #Total Below Poverty
    'S1701_C01_006E', #Male population
    'S1701_C02_006E', #Male Poverty
    'S1701_C01_007E', #Female population
    'S1701_C02_007E', #Female Poverty
    'S1701_C01_009E', #White Population
    'S1701_C02_009E', #White poverty
    'S1701_C01_010E', #Black Population
    'S1701_C02_010E', #Black Poverty
    'S1701_C01_011E', #American Indian And Alaskan Native Pop.
    'S1701_C02_011E', #American Indian And Alaskan Native Poverty
    'S1701_C01_012E', #Asian Population
    'S1701_C02_012E', #Asian Poverty
    'S1701_C01_013E', #Native Hawaiian and Other Pacific Islander Pop.
    'S1701_C02_013E', #Native Hawaiian and Other Pacific Islander poverty
    'S1701_C01_014E', #Other
    'S1701_C02_014E', #Other Poverty
    'S1701_C01_019E', #Population with less than high school education
    'S1701_C02_019E', #Below poverty level, less than high school education
    'S1701_C01_020E', #Population with high school graduate (includes equivalency)
    'S1701_C02_020E', #Below poverty level, high school graduate
    'S1701_C01_021E', #Population with some college or associate's degree
    'S1701_C02_021E', #Below poverty level, some college or associate's degree
    'S1701_C01_022E', #Population with bachelor's degree or higher
    'S1701_C02_022E'  #Below poverty level, bachelor's degree or higher
]
cleaned_dataframes = []
for df, year in datasets:
    # Keep only relevant columns
    cleaned_df = df[columns]

    # Drop first row
    cleaned_df = cleaned_df.drop(0)

    # Rename columns
    cleaned_df.columns = [
        'Geo_ID', 'Region', 'Total_Population',
        'Total_Below_Poverty', 'Male_Population', 'Poverty_Male',
        'Female_Population', 'Poverty_Female', 'White_Population',
        'Poverty_White', 'Black_Population', 'Poverty_Black',
        'AIAN_Population', 'Poverty_AIAN',
        'Asian_Population', 'Poverty_Asian', 'NHPI_Population',
        'Poverty_NHPI', 'Other_Population', 'Poverty_Other',
        'Population_Less_HS', 'Poverty_Less_HS',
        'Population_HS_Grad', 'Poverty_HS_Grad',
        'Population_Some_College', 'Poverty_Some_College',
        'Population_Bachelors_Higher', 'Poverty_Bachelors_Higher'
    ]

    # Add the year column
    cleaned_df['Year'] = year

    # Convert the necessary columns to numeric
    cleaned_df[
        ['Total_Population',
        'Total_Below_Poverty', 'Male_Population', 'Poverty_Male',
        'Female_Population', 'Poverty_Female', 'White_Population',
        'Poverty_White', 'Black_Population', 'Poverty_Black',
        'AIAN_Population', 'Poverty_AIAN',
        'Asian_Population', 'Poverty_Asian', 'NHPI_Population',
        'Poverty_NHPI', 'Other_Population', 'Poverty_Other',
        'Population_Less_HS', 'Poverty_Less_HS',
        'Population_HS_Grad', 'Poverty_HS_Grad',
        'Population_Some_College', 'Poverty_Some_College',
        'Population_Bachelors_Higher', 'Poverty_Bachelors_Higher']
    ] = cleaned_df[
          ['Total_Population',
        'Total_Below_Poverty', 'Male_Population', 'Poverty_Male',
        'Female_Population', 'Poverty_Female', 'White_Population',
        'Poverty_White', 'Black_Population', 'Poverty_Black',
        'AIAN_Population', 'Poverty_AIAN',
        'Asian_Population', 'Poverty_Asian', 'NHPI_Population',
        'Poverty_NHPI', 'Other_Population', 'Poverty_Other',
        'Population_Less_HS', 'Poverty_Less_HS',
        'Population_HS_Grad', 'Poverty_HS_Grad',
        'Population_Some_College', 'Poverty_Some_College',
        'Population_Bachelors_Higher', 'Poverty_Bachelors_Higher']
     ].apply(pd.to_numeric, errors='coerce')

    #Append the cleaned dataframe to the list
    cleaned_dataframes.append(cleaned_df)

#Combine all cleaned DataFrames into one
combined_df = pd.concat(cleaned_dataframes, ignore_index=True)


print(combined_df.head())

#saved to csv
combined_df.to_csv('combined_cleaned_poverty_data_2010_2023_with_year.csv', index=False)

```
